```python
import pandas as pd

data = pd.read_csv("5-data.csv")
data
```

```python
total_records = len(data)
prob = lambda X:X/total_records
print(f"Total number of records: {total_records}")
```

```python
golf_recreation = len(data.where(data["Recreation"] == "golf").dropna())
prob_golf = prob(golf_recreation)
print(f"""Number of golf recreation records: {golf_recreation}\nUnconditional probability of golf: {prob_golf*100} %""")
```

```python
med_risk_single = len(data.where(data["Worthiness"] == "med-risk").where(data["Status"] == "single").dropna())
med_risk = len(data.where(data["Worthiness"] == "med-risk").dropna())

cond_prob = prob(med_risk_single)/prob(med_risk)

print(f"""Number of med-risk worthiness and single status records: {med_risk_single}
Number of med-risk worthiness records: {med_risk}
Conditional probability of single given med-risk: {cond_prob*100} %""")
```
